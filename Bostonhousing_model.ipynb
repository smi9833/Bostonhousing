```python
m=lr.coef_
print("Slopr: ",m)
c=lr.intercept_
print("Intercept: ",c)

#coef_ and intercept_ both are inbuilt method of LineaRegression class
```

```python
X_test.head()
```

```python
#Test the model : use inbuilt method predict of LinearRegression class
#equation of straight line y_predict=m*X+c
Y_pred=lr.predict(X_test)
```

```python
Y_pred
```

```python
#Evaluate of model
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(Y_test,Y_pred)
```

```python
#Evalution of model : find r2-score means r-squared (0-1)
#r2-score : check model is best perform or not
#r2_score inbuilt class which define in sklearn.metrics
from sklearn.metrics import r2_score
```

```python
r2_score(Y_test,Y_pred)
```

```python
#here r2-score is good because nearest to +1 and MSE is very low means we
#create a successful model
```

# 1.Multicolinearity

```python
plt.figure(figsize=(12,8))
sns.heatmap(df.corr(),annot=True)
plt.show()
```

# Observation:
    we select those features which have a high correlation with our target variable Price. By looking at the correlation we can see that RM has a strong positive correlation with price (0.7) ,LSTAT has a high negative correlation with price(-0.74).
    and we also check for multicolinearity the fetures rad, tax, dis, age are highly correlated to each other.

```python
X=df[["rm","lstat"]]#input variable
Y=df["price"]#output variable
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3,random_state=1)
```

```python
lr1=LinearRegression()
```

```python
#train the model using fit method
lr1.fit(X_train,Y_train)
```

```python
m=lr1.coef_
print("slope:",m)
c=lr1.intercept_
print("Intercept:",c)
```

```python
#test the model
Y_pred=lr1.predict(X_test)
```

```python
#Evaluate model
mean_squared_error(Y_test,Y_pred)
```

```python
r2_score(Y_test,Y_pred)
```

```python
#multicolinearity is solved
#score is 72% is good
```
